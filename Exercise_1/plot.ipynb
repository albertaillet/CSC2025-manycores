```python
def S(N: int, P: float) -> float:
    return 1 / ((1 - P) + P / N)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a range of N values (number of processors)
N_values = np.arange(1, 101, 1)

# Define different values of P (proportion of parallelizable part)
P_values = [0.5, 0.7, 0.9, 0.95]

# Plot Amdahl's Law for different values of P
plt.figure(figsize=(10, 6))
for P in P_values:
    S_values = [S(N, P) for N in N_values]
    plt.plot(N_values, S_values, label=f'P={P}')

plt.title("Amdahl's Law")
plt.xlabel("Number of Processors (N)")
plt.ylabel("Speedup (S(N))")
plt.legend()
plt.grid(True)
plt.show()
```

Suppose now to be in charge of the hardware resources of a small manufactur-
ing company. The main quad-core web-server works OK, but the performance
of your web application running on it is suboptimal. 75% of the total work done
by the program is spent in I/O, which runs purely sequentially. The rest of the
operations scale well with the number of available cores. What is the gain you
expect moving to an eight-core server? And to a hypothetical machine with 200
cores?

```python
P = (1 - 0.75)
print("speedups:", S(8, P), S(200, P))
```
